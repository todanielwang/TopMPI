```python
import pandas as pd
import math

r1A = pd.read_csv("switchPrecursor/1A.tsv", delimiter="\t")
r1B = pd.read_csv("switchPrecursor/1B.tsv", delimiter="\t")
r2A = pd.read_csv("switchPrecursor/2A.tsv", delimiter="\t")
r2B = pd.read_csv("switchPrecursor/2B.tsv", delimiter="\t")
```

```python
r1A = r1A[["Scan(s)", "Proteoform", "Proteoform mass"]]
r1B = r1B[["Scan(s)", "Proteoform", "Proteoform mass"]]
r2A = r2A[["Scan(s)", "Proteoform", "Proteoform mass"]]
r2B = r2B[["Scan(s)", "Proteoform", "Proteoform mass"]]
```

```python
r1 = r1A.merge(r1B, on="Scan(s)", how="outer")
r2 = r2A.merge(r2B, on="Scan(s)", how="outer")
```
